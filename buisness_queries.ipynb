# 6. Calculate the number of orders per month in 2018.

```python
df=pd.read_sql_query("""
    SELECT 
        CASE strftime('%m', order_purchase_timestamp)
            WHEN '01' THEN 'Jan'
            WHEN '02' THEN 'Feb'
            WHEN '03' THEN 'Mar'
            WHEN '04' THEN 'Apr'
            WHEN '05' THEN 'May'
            WHEN '06' THEN 'Jun'
            WHEN '07' THEN 'Jul'
            WHEN '08' THEN 'Aug'
            WHEN '09' THEN 'Sep'
            WHEN '10' THEN 'Oct'
            WHEN '11' THEN 'Nov'
            WHEN '12' THEN 'Dec'
        END AS month_name,
        COUNT(*) AS number_of_orders
    FROM orders
    WHERE strftime('%Y', order_purchase_timestamp) = '2018'
    GROUP BY strftime('%m', order_purchase_timestamp)
    ORDER BY strftime('%m', order_purchase_timestamp)
""", conn)

plt.figure(figsize=(12, 6)) 

sns.barplot(x=df["month_name"],y=df["number_of_orders"],color=None)
plt.title('Number of Orders per Month in 2018', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)

for i, v in enumerate(df["number_of_orders"]):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.grid(False)
plt.show()
```

# 7. Find the average number of products per order, grouped by customer city.

```python
pd.read_sql_query("""with count_per_order as(select orders.order_id,orders.customer_id,count(order_items.order_item_id) as oc
    from orders join order_items
    on orders.order_id=order_items.order_id
    group by orders.order_id,orders.customer_id)
    select customers.customer_city,round(avg(count_per_order.oc),2) as avg_order
    from customers join count_per_order
    on customers.customer_id=count_per_order.customer_id
    group by customers.customer_city
    order by avg_order desc
    """,conn)
```

# 8. Calculate the percentage of total revenue contributed by each product category.

# 10. Calculate the total revenue generated by each seller, and rank them by revenue.

# 11. Calculate the moving average of order values for each customer over their order history.

```python
pd.read_sql_query("""select customer_id,order_purchase_timestamp,payment,
    avg(payment) over(partition by customer_id order by order_purchase_timestamp
    rows between 2 preceding and current row) as moving_avg
    from
    (select orders.customer_id,orders.order_purchase_timestamp,payments.payment_value as payment
    from payments join orders
    on payments.order_id=orders.order_id) as a""",conn)
```

# 12. Calculate the cumulative sales per month for each year.

```python
pd.read_sql_query("""SELECT 
    strftime('%Y', orders.order_purchase_timestamp) AS sale_year,
    strftime('%m', orders.order_purchase_timestamp) AS sale_month,
    SUM(payments.payment_value) AS monthly_sales,
    SUM(SUM(payments.payment_value)) OVER (
        PARTITION BY strftime('%Y', orders.order_purchase_timestamp) 
        ORDER BY strftime('%m', orders.order_purchase_timestamp)
        ROWS UNBOUNDED PRECEDING
    ) AS cumulative_monthly_sales
FROM orders join payments
on orders.order_id =payments.order_id
GROUP BY strftime('%Y', orders.order_purchase_timestamp), strftime('%m', orders.order_purchase_timestamp)
ORDER BY sale_year, sale_month""",conn)
```

# 13. Calculate the year-over-year growth rate of total sales.

```python
pd.read_sql_query("""with a as(SELECT 
    strftime('%Y', orders.order_purchase_timestamp) AS year,
    round(SUM(payments.payment_value),2) AS sales 
    from orders join payments
    on orders.order_id=payments.order_id
    group BY year
    order by year)
    select year,((sales-lag(sales,1)over(order by year))/lag(sales,1)over(order by year))*100 as YoY_growth 
    from a""",conn)
        
```

# 14. Calculate the retention rate of customers, defined as the percentage of customers who make another purchase within 6 months of their first purchase.

```python
pd.read_sql_query("""
    WITH a AS (
        SELECT 
            customers.customer_id,
            MIN(orders.order_purchase_timestamp) AS first_order
        FROM customers 
        JOIN orders ON customers.customer_id = orders.customer_id
        GROUP BY customers.customer_id
    ),
    b AS (
        SELECT 
            a.customer_id, 
            COUNT(DISTINCT orders.order_purchase_timestamp) AS next_order
        FROM a 
        JOIN orders ON orders.customer_id = a.customer_id
            AND orders.order_purchase_timestamp > a.first_order
            AND orders.order_purchase_timestamp < DATE(a.first_order, '+6 months')
        GROUP BY a.customer_id
    ) 
    SELECT 
        100.0 * COUNT(DISTINCT b.customer_id) / COUNT(DISTINCT a.customer_id) AS percentage
    FROM a 
    LEFT JOIN b ON a.customer_id = b.customer_id
""", conn)
```

# 15. Identify the top 3 customers who spent the most money in each year.

```python
df=pd.read_sql_query("""
    SELECT years, customer_id, payment, d_rank
    FROM (
        SELECT 
            CAST(strftime('%Y', orders.order_purchase_timestamp) AS INTEGER) AS years,
            orders.customer_id,
            SUM(payments.payment_value) AS payment,
            DENSE_RANK() OVER (
                PARTITION BY CAST(strftime('%Y', orders.order_purchase_timestamp) AS INTEGER)
                ORDER BY SUM(payments.payment_value) DESC
            ) AS d_rank
        FROM orders 
        JOIN payments ON payments.order_id = orders.order_id
        GROUP BY 
            CAST(strftime('%Y', orders.order_purchase_timestamp) AS INTEGER),
            orders.customer_id
    ) AS a
    WHERE d_rank <= 3
""", conn)

plt.figure(figsize=[12,6])
sns.barplot(x="customer_id", y="payment", data=df, hue="years")
plt.xticks(rotation=90)
plt.title('Top 3 Customers by Payment per Year')
plt.xlabel('Customer ID')
plt.ylabel('Total Payment')
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```
